# Task
Train a simple machine learning model using synthetic data.

## Generate synthetic data

### Subtask:
Create a simple dataset for a regression problem using scikit-learn's `make_regression` function.


**Reasoning**:
Generate a synthetic dataset for regression using `make_regression` as instructed.



```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, n_targets=1, random_state=42)
```

## Split data

### Subtask:
Split the synthetic data into training and testing sets.


**Reasoning**:
Split the synthetic data into training and testing sets using `train_test_split`.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Define and train model

### Subtask:
Define a simple linear regression model using scikit-learn and train it on the training data.


**Reasoning**:
Define and train a simple linear regression model using scikit-learn.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

## Evaluate model

### Subtask:
Evaluate the trained model on the testing data using a relevant metric like Mean Squared Error.


**Reasoning**:
Evaluate the trained model on the testing data using Mean Squared Error.



```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

## Summary:

### Data Analysis Key Findings

*   A synthetic regression dataset with 1000 samples and 10 features was successfully generated.
*   The dataset was split into training and testing sets, with 75% allocated for training and 25% for testing.
*   A Linear Regression model was defined and trained on the synthetic training data.
*   The trained model was evaluated on the testing data, resulting in a Mean Squared Error (MSE) of approximately 2.31e-27.

### Insights or Next Steps

*   The extremely low MSE suggests that the Linear Regression model is a very good fit for this synthetic data, likely because the data was generated with a linear relationship.
*   For a more realistic evaluation, the model could be tested on more complex synthetic data or real-world datasets.
